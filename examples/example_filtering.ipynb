```python
%load_ext autoreload
%autoreload 2
%matplotlib notebook

import matplotlib.pyplot as plt
import imageio

# Low-level operators (backend independent)
import llops as yp

import llops.filter as filters

yp.setDefaultBackend('numpy')
```

```python
x = yp.simulation.brain()
plt.figure(figsize=(5,3))
plt.imshow(yp.real(x))
```

## Gallery of Filters (Real Object)

```python
image = x

plt.figure(figsize=(10,10))
plt.subplot(351)
plt.imshow(yp.abs(image))
plt.title('Original')
plt.subplot(352)
plt.imshow(yp.abs(filters.sobel(image)))
plt.title('Sobel Filter')
plt.subplot(353)
plt.imshow(yp.abs(filters.canny(image)))
plt.title('Canny Filter')
plt.subplot(354)
plt.imshow(yp.abs(filters.roberts(image)))
plt.title('Roberts Filter')
plt.subplot(355)
plt.imshow(yp.abs(filters.prewitt(image)))
plt.title('Prewitt Filter')
plt.subplot(356)
plt.imshow(yp.abs(filters.scharr(image)))
plt.title('Scharr Filter')
plt.subplot(357)
plt.imshow(yp.abs(filters.laplace(image)))
plt.title('Laplace Filter')
plt.subplot(358)
plt.imshow(yp.abs(filters.gaussian(image)))
plt.title('Gaussian Filter')
plt.subplot(359)
plt.imshow(yp.abs(filters.median(image)))
plt.title('Median Filter')

# plt.subplot(3,5,12)
# plt.imshow(yp.abs(filters.bilateral(image)))
# plt.title('Bilateral Filter')
```

## Gallary of Filters, Real Object, Arrayfire

```python
image = yp.changeBackend(x, 'arrayfire')

plt.figure(figsize=(10,10))
plt.subplot(351)
plt.imshow(yp.abs(image))
plt.title('Original')
plt.subplot(352)
plt.imshow(yp.abs(filters.sobel(image)))
plt.title('Sobel Filter')
plt.subplot(353)
plt.imshow(yp.abs(filters.canny(image)))
plt.title('Canny Filter')
plt.subplot(354)
plt.imshow(yp.abs(filters.roberts(image)))
plt.title('Roberts Filter')
plt.subplot(355)
plt.imshow(yp.abs(filters.prewitt(image)))
plt.title('Prewitt Filter')
plt.subplot(356)
plt.imshow(yp.abs(filters.scharr(image)))
plt.title('Scharr Filter')
plt.subplot(357)
plt.imshow(yp.abs(filters.laplace(image)))
plt.title('Laplace Filter')
plt.subplot(358)
plt.imshow(yp.abs(filters.gaussian(image)))
plt.title('Gaussian Filter')
plt.subplot(359)
plt.imshow(yp.abs(filters.median(image)))
plt.title('Median Filter')

# plt.subplot(3,5,12)
# plt.imshow(yp.abs(filters.bilateral(image)))
# plt.title('Bilateral Filter')
```

```python
plt.figure()
plt.subplot(121)
plt.imshow(yp.abs(filters._gaussianKernel((100,100), 10, backend='arrayfire')))
plt.colorbar()

plt.subplot(122)
plt.imshow(yp.abs(filters._gaussianKernel((100,100), 10, backend='numpy')))
plt.colorbar()

```

```python
x_np = yp.randn((100,100))
x_af = yp.changeBackend(x_np, 'arrayfire')

sigma = 1
y_np = yp.filter.canny(x_np, sigma=sigma)
y_af = yp.filter.canny(x_af, sigma=sigma)

plt.figure()
plt.subplot(131)
plt.imshow(yp.abs(yp.changeBackend(y_af, 'numpy')))
plt.subplot(132)
plt.imshow(yp.abs(y_np))
plt.subplot(133)
plt.imshow(yp.abs(y_np - yp.changeBackend(y_af, 'numpy')))
plt.colorbar()
```
