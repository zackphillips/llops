```python
%load_ext autoreload
%autoreload 2
%matplotlib notebook

import llops.simulation as sim
import matplotlib.pyplot as plt
import llops as yp
import math
```

## Gallery of Available Test Objects

```python
# Get list of available objects
test_object_list = sim.get_available_objects()

x_size = 5
y_size = math.ceil(len(test_object_list) / x_size)

plt.figure(figsize=(12, 3 * y_size))

for index, label in enumerate(test_object_list):
    test_object = sim.testObject(label)
    
    plt.subplot(y_size, x_size, index + 1)
    plt.imshow(yp.real(test_object))
    plt.title(label)
```
