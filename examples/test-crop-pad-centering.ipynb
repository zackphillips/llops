## Test Centered Pad and FFT

```python
# Define a image with an odd number of pixels
img_odd = yp.zeros((101,101))
img_odd[51,51] = 1

img_odd_pad_even_size = (200,200)
img_odd_pad_even = yp.pad(img_odd, img_odd_pad_even_size, center=True)

img_odd_pad_odd_size = (201,201)
img_odd_pad_odd = yp.pad(img_odd, img_odd_pad_odd_size, center=True)

img_even = yp.zeros((100,100))
img_even[50,50] = 1

img_even_pad_even_size = (200,200)
img_even_pad_even = yp.pad(img_even, img_even_pad_even_size, center=True)

img_even_pad_odd_size = (200,200)
img_even_pad_odd = yp.pad(img_even, img_even_pad_odd_size, center=True)

plt.figure()
plt.subplot(231)
plt.imshow(yp.angle(yp.Ft(img_odd)))
plt.subplot(232)
plt.imshow(yp.angle(yp.Ft(img_odd_pad_odd)))
plt.subplot(233)
plt.imshow(yp.angle(yp.Ft(img_odd_pad_even)))
plt.subplot(234)
plt.imshow(yp.angle(yp.Ft(img_even)))
plt.subplot(235)
plt.imshow(yp.angle(yp.Ft(img_even_pad_odd)))
plt.subplot(236)
plt.imshow(yp.angle(yp.Ft(img_even_pad_even)))
```

# Ensure Equality When Cropping

```python
# Define a image with an odd number of pixels
img_odd = yp.zeros((101,101))
img_odd[50,50] = 1

img_odd_pad_even_size = (200,200)
img_odd_pad_even = yp.pad(img_odd, img_odd_pad_even_size, center=True)
img_odd_pad_even_crop = yp.crop(img_odd_pad_even, yp.shape(img_odd), center=True)
yp.assert_equality(img_odd_pad_even_crop, img_odd)

img_odd_pad_odd_size = (201,201)
img_odd_pad_odd = yp.pad(img_odd, img_odd_pad_odd_size, center=True)
img_odd_pad_odd_crop = yp.crop(img_odd_pad_odd, yp.shape(img_odd), center=True)
yp.assert_equality(img_odd_pad_odd_crop, img_odd)

img_even = yp.zeros((100,100))
img_even[50,50] = 1

img_even_pad_even_size = (200,200)
img_even_pad_even = yp.pad(img_even, img_even_pad_even_size, center=True)
img_even_pad_even_crop = yp.crop(img_even_pad_even, yp.shape(img_even), center=True)
yp.assert_equality(img_even_pad_even_crop, img_even)

img_even_pad_odd_size = (200,200)
img_even_pad_odd = yp.pad(img_even, img_even_pad_odd_size, center=True)
img_even_pad_odd_crop = yp.crop(img_even_pad_odd, yp.shape(img_even), center=True)
yp.assert_equality(img_even_pad_odd_crop, img_even)
```
